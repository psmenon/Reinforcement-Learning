### Part 1: TD Control: Sarsa

In this section, you will write your own implementation of the Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
# Additional functions for SARSA

def generate_policy(Q,state,nA,epsilon):
    policy = np.ones(nA) * epsilon / nA
    best_action = np.argmax(Q[state])
    policy[best_action] = 1 - epsilon + (epsilon / nA)
    return policy

def generate_episode_Sarsa(env,Q,nA,epsilon,alpha,gamma):
    state = env.reset()
    p = generate_policy(Q,state,nA,epsilon)
    action = np.random.choice(np.arange(nA),p=p) \
                                                  if state in Q else env.action_space.sample()
    while True:
        new_state,reward,done,info = env.step(action)
        p = generate_policy(Q,new_state,nA,epsilon)
        action_ = np.random.choice(np.arange(nA),p=p) \
                                                      if state in Q else env.action_space.sample()
        
        Q[state][action] += alpha * (reward + (gamma * Q[new_state][action_]) - Q[state][action])
        state = new_state
        action = action_
        
        if done:
            break
    return Q
        
````

````python
def sarsa(env, num_episodes, alpha, gamma=1.0):
    
    nA = env.nA
    # initialize action-value function (empty dictionary of arrays)
    Q = defaultdict(lambda: np.zeros(env.nA))
    
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()  
        
        epsilon = 1 / i_episode
        Q = generate_episode_Sarsa(env,Q,nA,epsilon,alpha,gamma)
        
    return Q
````

### Part 2: TD Control: Q-learning

In this section, you will write your own implementation of the Q-learning control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
# Additional functions for Q-learning

def generate_policy_Q_learning(Q,State,epsilon,nA):
    policy = np.ones(nA) * epsilon / nA
    best_action = np.argmax(Q[State])
    policy[best_action] = 1 - epsilon + (epsilon / nA)
    return policy

def generate_episode_epsilon_greedy_Q_learning(env,Q,nA,epsilon,gamma,alpha):
    state = env.reset()
    while True:
        p = generate_policy_Q_learning(Q,state,epsilon,nA)
        action = np.random.choice(np.arange(nA),p=p) \
                                                      if state in Q else env.action_space.sample()
        
        new_state,reward,done,info = env.step(action)
        Q[state][action] += alpha * (reward + (gamma * Q[new_state][np.argmax(Q[state])]) - Q[state][action])
        state = new_state
        if done:
            break
    return Q
````

````python
def q_learning(env, num_episodes, alpha, gamma=1.0):
    nA = env.nA
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(env.nA))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        epsilon = 1 / i_episode
        Q = generate_episode_epsilon_greedy_Q_learning(env,Q,nA,epsilon,gamma,alpha)
        
    return Q
````

### Part 3: TD Control: Expected Sarsa

In this section, you will write your own implementation of the Expected Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
# Additional functions for expected SARSA

def generate_policy_expected_sarsa(Q,state,epsilon,nA):
    policy = np.ones(nA) * epsilon / nA
    best_action = np.argmax(Q[state])
    policy[best_action] = 1 - epsilon + (epsilon / nA)
    return policy

def generate_episode_expected_sarsa(env,Q,nA,epsilon,gamma,alpha):
    state = env.reset()
    while True:
        p = generate_policy_expected_sarsa(Q,state,epsilon,nA)
        action = np.random.choice(np.arange(nA),p=p) \
                                                      if state in Q else env.action_space.sample()
        
        new_state,reward,done,info = env.step(action)
        
        p_ = generate_policy_expected_sarsa(Q,new_state,epsilon,nA)
        sum_ = np.dot(p_,Q[new_state])
        
        Q[state][action] += alpha * (reward + (gamma * sum_) - Q[state][action])
        state = new_state
        
        if done:
            break
    return Q
````

````python
def expected_sarsa(env, num_episodes, alpha, gamma=1.0):
    nA = env.nA
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(env.nA))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        epsilon = 0.005
        Q = generate_episode_expected_sarsa(env,Q,nA,epsilon,gamma,alpha)
        
    return Q
````
